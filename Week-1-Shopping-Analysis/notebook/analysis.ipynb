```python
# Shopping Dataset Analysis
# Internship Week 1 Assignment

# Importing required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Loading dataset

df = pd.read_csv("../data/combined_dataset.csv")

# Display first 5 rows

df.head()
```

```python
# Selecting important columns

selected_columns = df[['title', 'category', 'rating', 'final_price']]

selected_columns.head()
```

```python
# Filtering products with rating greater than 4

high_rated_products = df[df['rating'] > 4]

high_rated_products.head()
```

```python
# Filling missing values in discount column using mean

df['discount'] = df['discount'].fillna(df['discount'].mean())
df.isnull().sum()
```

```python
# Checking duplicate rows

print("Duplicate Rows:", df.duplicated().sum())
```

```python
# Converting final_price to numeric format

df['final_price'] = df['final_price'].replace('[₹,"]', '', regex=True).astype(float)

# Creating price difference column

df['price_difference'] = df['initial_price'] - df['final_price']

# Creating popularity metric

df['popularity'] = df['rating'] * df['ratings_count']

# Creating quantity column

df['quantity'] = np.random.randint(1, 5, size=len(df))

# Creating total amount column

df['total_amount'] = df['final_price'] * df['quantity']

# Saving cleaned dataset

df.to_csv("../cleaned_data/cleaned_dataset.csv", index=False)
```

# Key Insights

1. The dataset contains 1000 products across multiple categories.
2. Missing values in the discount column were handled using mean imputation.
3. Duplicate records were checked and removed.
4. New features like price difference and popularity were created.
5. The cleaned dataset was successfully exported.
